```python
import pandas as pd
import numpy as np
```

```python
df_v = pd.read_csv('../../data/processed/v.csv')
df_v.drop(columns=  ['Unnamed: 0', 'categoria_tipo', 'empresa_modalidade', 'veiculo_validador', 'passageiro_evento', 'movimentos_data_arq',
                    'passageiro_integracao_bum', 'passageiro_valor_subsidio', 'passageiro_sigben', 'passageiro_valor_repasse_metro'], inplace=True)

colunas_date = ['movimento_diario_data_mov', 'linha_hora_abertura', 'passageiro_data_hora',
                'linha_hora_fechamento', 'viagem_data_hora_abertura', 'viagem_data_hora_fechamento']
for i in colunas_date:
    df_v[i]= pd.to_datetime(df_v[i], infer_datetime_format=True)

df_v.head()
```

```python
df_v = df_v[(df_v.movimento_diario_data_mov >= pd.to_datetime('2021-09-01')) & (df_v.movimento_diario_data_mov <= pd.to_datetime('2021-09-07'))]
df_v.shape
```
